# image_detection using matplotlib,opencv,mtcnn


```python
## importing libraries
%matplotlib inline
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
import cv2

def show_image(filename, result_list,frame):
    data = pyplot.imread(filename)  ## loads the image
    pyplot.imshow(data) ## plotting the image
    ax = pyplot.gca() # Object creation for drawing shapes
    # plot each box
    
    for result in result_list:
        x, y, width, height = result['box'] ## retrieving coordinates of face
        rect = Rectangle((x, y), width, height, fill=False, color='red')  # creating the rectangular shape of face
        ax.add_patch(rect) ##drawing rectangular box
        for key, value in result['keypoints'].items():
            print(key,value)
            # create and draw dot
            if key=='left_eye' or key=='right_eye':
                dot = Circle(value, radius=8, color='red') # creating the circular shape of eyes
                ax.add_patch(dot)##drawing circular shape on image
            if key=='nose':
                dot = Circle(value, radius=4, color='red') # creating the circular shape of nose
                ax.add_patch(dot)##drawing circular shape on image
            if key=='mouth_left' or key=='mouth_right':
                dot = Circle(value,radius=4, color='red') # creating the circular shape of mouth
                ax.add_patch(dot)##drawing circular shape on image
                
    pyplot.axis('off')#show the plot
    pyplot.show()

    for i in range(len(result_list)):
        
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        
        pyplot.subplot(1, len(result_list), i+1)
        pyplot.axis('off')
        pyplot.imshow(data[y1:y2, x1:x2])
    pyplot.show()

inp=input("Enter '1' for Video and '2' for detecting static image file")
if inp=='1':
    vid=input("Enter '1' for detectig face through Webcam and '2' for detecting face in video")
    if vid=='1':
        cap = cv2.VideoCapture(0)
    else:
        video_file=input("Enter name of the video file with extension and should be present in same directory")
        cap = cv2.VideoCapture(video_file)
    detector = MTCNN()
    while 1:
        ret, frame = cap.read()
        result = detector.detect_faces(frame)
        if result != []:
            for person in result: ## drawing rectangular and circular shapes in order to depict the facial features
                bounding_box = person['box']
                keypoints = person['keypoints']
                cv2.rectangle(frame,(bounding_box[0], bounding_box[1]),(bounding_box[0]+bounding_box[2], bounding_box[1] + bounding_box[3]),
                (0,155,255), 2)
                cv2.rectangle(frame,(keypoints['mouth_left']),(keypoints['mouth_left'][0]+5,keypoints['mouth_left'][1]+5), (0,155,255), 2)
                cv2.rectangle(frame,(keypoints['mouth_right']),(keypoints['mouth_right'][0]+5,keypoints['mouth_right'][1]+5), (0,155,255), 2)
                cv2.circle(frame,(keypoints['nose']), 3, (0,155,255), 2)
                cv2.circle(frame,(keypoints['left_eye']), 8, (0,155,255), 2)
                cv2.circle(frame,(keypoints['right_eye']), 8, (0,155,255), 2)
            for i in range(len(result)):
                data=frame
        
                x1, y1, width, height = result[i]['box']
                x2, y2 = x1 + width, y1 + height

                pyplot.subplot(1, len(result), i+1)
                pyplot.axis('off')
                pyplot.imshow(data[y1:y2, x1:x2])
        pyplot.show()
      
        #display resulting frame
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) &0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

else:
    filename = input('Enter Filename of the image with extension and should be present in same directory')
    pixels = cv2.imread(filename)
    detector = MTCNN()
    faces = detector.detect_faces(pixels)
    show_image(filename, faces,pixels)
```
